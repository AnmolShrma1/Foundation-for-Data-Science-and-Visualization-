```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your data
df = pd.read_csv("D:\DataScience Lab\data.csv")  # Update this path to the actual location of your CSV file

# Check the column names to make sure they are correct
print(df.columns)

# Display a color palette
sns.palplot(sns.color_palette("magma", 10))

# Plotting with Seaborn scatterplot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Duration', y='Calories', palette='magma', hue='Calories', legend=None)

# Style and display
sns.set_style("ticks")
sns.despine()
plt.xlabel('Duration')
plt.ylabel('Calories')
plt.title('Scatter plot of Duration vs. Calories')
plt.show()
```
